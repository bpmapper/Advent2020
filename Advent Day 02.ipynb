# Day 2 - Part 1 Invalid passwords

```python
# read in data
import pandas as pd
df=pd.read_csv('day2.txt', header=None)
df.sample(5)
```

```python
# split into 3 columns sep by space
df[["a", "b", "pword"]] = df.iloc[:,0].str.split(pat=" ", expand=True)
df.sample(5)
```

```python
# Extract search letter and hi/low
df['letter']=df.b.str.replace(":","")
df[["low", "hi"]] = df['a'].str.split(pat="-", expand=True)
df['hi']=pd.to_numeric(df.hi)
df['low']=pd.to_numeric(df.low)
df.sample(5)
```

```python
df['freq']=df.apply(lambda x: x.pword.count(x.letter), axis=1)
df['result']=df.apply(lambda x: ("OK" if x.freq >= x.low and x.freq<= x.hi else "Fail") , axis=1)
df.sample(5)
```

```python
df.result.value_counts()
```

# Part 2 - Positions

```python
df['pos1']=df.apply(lambda x: x.pword[x.low-1:x.low], axis=1)
df['pos2']=df.apply(lambda x: x.pword[x.hi-1:x.hi], axis=1)
df['result2']=df.apply(lambda x: ("OK" if (x.pos1==x.letter or x.pos2==x.letter) and x.pos1!=x.pos2 else "Fail"), axis=1)
df
```
